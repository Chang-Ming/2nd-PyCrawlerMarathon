### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")
text = soup.find(class_='part_list_2').find_all('h3')
for d in text:
    print(d.find(class_='date').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news = {}
for d in text:
    date = d.find(class_='date').text
    title = d.find_all('a')[-1].text
    tag = d.find('em').text
    news.setdefault(tag, [])
    news[tag].append({'date':date, 'title':title})
print(news)
```

### ③ 哪一個類別的文章最多

```python
number ={}
for tag in list(news.keys()):
    number[tag] = len(news[tag])
number
```

```python
m = max(number, key = number.get)
print(m, number[m])
```
