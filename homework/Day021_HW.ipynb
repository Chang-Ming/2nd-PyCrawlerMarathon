# Ettoday 網路爬蟲實作練習


* 能夠利用 Request + BeatifulSour 撰寫爬蟲，並存放到合適的資料結構


## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time
import pandas as pd
```

## 1. 取出今天所有的新聞

```python
# 取得昨天時間
date_time = datetime.now() - timedelta(1)
date_time = date_time.strftime("%Y/%m/%d")
print("取得昨天時間：", date_time)
```

```python
executable_path = "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe" #chromedriver.exe執行檔所存在的路徑
browser = webdriver.Chrome(executable_path)
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break
```

```python
# 取得資料，丟到 BeautifulSoup 解析
news_day = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    if date_time not in d.find(class_="date").text:
        news_day.append({'Date' : d.find(class_="date").text, 'Title' : d.find_all('a')[-1].text})
pd.DataFrame(news_day)
```

```python
browser.close()
```

## 2. 取出現在時間兩小時內的新聞

```python
# 取得昨天時間
date_time = datetime.now() - timedelta(hours = 2)
date_time = date_time.strftime("%Y/%m/%d %H:%M")
print("取得兩小時前的時間：", date_time)
```

```python
executable_path = "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe" #chromedriver.exe執行檔所存在的路徑
browser = webdriver.Chrome(executable_path)
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if date_time > soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break
```

```python
# 取得資料，丟到 BeautifulSoup 解析
news_2hours = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    if date_time <= d.find(class_="date").text:
        news_2hours.append({'Date' : d.find(class_="date").text, 'Title' : d.find_all('a')[-1].text})
pd.DataFrame(news_2hours)
```

```python
browser.close()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
date_time = datetime.now() - timedelta(3)
date_time = date_time.strftime("%Y/%m/%d")
print("取得昨天時間：", date_time)
```

```python
# 取得昨天時間
date_time = datetime.now() - timedelta(3)
date_time_three = date_time.strftime("%Y/%m/%d")
date_time_month = str(int(date_time.strftime("%m")))
date_time_day = str(int(date_time.strftime("%d")))
three_days_15_clock = datetime.strptime(date_time_three + ' 15', '%Y/%m/%d %H').strftime('%Y/%m/%d %H:%M:%S')
three_days_17_clock = datetime.strptime(date_time_three + ' 17', '%Y/%m/%d %H').strftime('%Y/%m/%d %H:%M:%S')
print("取得三天前時間：", date_time.strftime("%Y/%m/%d"))
print("取得三天前時間：", three_days_15_clock)
print("取得三天前時間：", three_days_17_clock)
```

```python
executable_path = "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe" #chromedriver.exe執行檔所存在的路徑
browser = webdriver.Chrome(executable_path)
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 選擇且點選日期

selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_time_month)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_time_day)
browser.find_element_by_id('button').click()
```

```python
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text < three_days_15_clock:
        print('==== STOP ====')
        break
```

```python
# 取得資料，丟到 BeautifulSoup 解析
news_three_day = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    if three_days_15_clock < d.find(class_="date").text < three_days_17_clock:
        news_three_day.append({'Date' : d.find(class_="date").text, 'Title' : d.find_all('a')[-1].text})
pd.DataFrame(news_three_day)
```

```python
browser.close()
```
